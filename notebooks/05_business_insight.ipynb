```python
# =========================================================
# STEP 5 — AI STRATEGIC INSIGHT ENGINE
# FILE: notebooks/05_business_insight.ipynb
# =========================================================

# =========================================================
# STEP 5.1 — IMPORT LIBRARY
# =========================================================

import pandas as pd
import numpy as np
import warnings

import plotly.express as px

warnings.filterwarnings("ignore")


# =========================================================
# STEP 5.2 — LOAD CLEAN DATASET
# =========================================================

DATA_PATH = "../data/processed/clean_public_sentiment.csv"

df = pd.read_csv(DATA_PATH, parse_dates=["created_at"])

print("Dataset Shape:", df.shape)
df.head()


# =========================================================
# STEP 5.3 — EXECUTIVE KPI CALCULATION
# =========================================================

total_comments = len(df)

positive_rate = (df["sentiment"] == "Positive").mean() * 100
neutral_rate = (df["sentiment"] == "Neutral").mean() * 100
negative_rate = (df["sentiment"] == "Negative").mean() * 100

avg_engagement = df["engagement_score"].mean()

high_urgency_rate = (df["urgency_level"] == "High").mean() * 100

top_program = (
    df.groupby("program_name")["engagement_score"]
      .sum()
      .sort_values(ascending=False)
      .index[0]
)

top_hashtag = (
    df.groupby("hashtags")["engagement_score"]
      .sum()
      .sort_values(ascending=False)
      .index[0]
)

print("Total Comments      :", total_comments)
print("Positive Rate       :", round(positive_rate, 2), "%")
print("Negative Rate       :", round(negative_rate, 2), "%")
print("Average Engagement  :", round(avg_engagement, 2))
print("High Urgency Rate   :", round(high_urgency_rate, 2), "%")
print("Top Program         :", top_program)
print("Top Hashtag         :", top_hashtag)
```

```python
# =========================================================
# STEP 5.4 — PUBLIC TRUST INDEX
# =========================================================

# Formula sederhana:
# Positive Sentiment - Negative Sentiment + 100
# Rentang hasil akan berada sekitar 0–200.
# Semakin tinggi, semakin baik persepsi publik.

public_trust_index = 100 + positive_rate - negative_rate

print(f"Public Trust Index: {public_trust_index:.2f}")
```

```python
# =========================================================
# STEP 5.5 — GOVERNMENT RESPONSE STATUS
# =========================================================

if public_trust_index >= 140:
    response_status = "Excellent"
elif public_trust_index >= 120:
    response_status = "Strong"
elif public_trust_index >= 100:
    response_status = "Stable"
else:
    response_status = "Needs Attention"

print("Government Response Status:", response_status)
```

```python
# =========================================================
# STEP 5.6 — KPI VISUALIZATION
# =========================================================

kpi_df = pd.DataFrame({
    "Metric": [
        "Positive Sentiment",
        "Negative Sentiment",
        "High Urgency",
        "Public Trust Index"
    ],
    "Value": [
        positive_rate,
        negative_rate,
        high_urgency_rate,
        public_trust_index
    ]
})

fig = px.bar(
    kpi_df,
    x="Metric",
    y="Value",
    color="Value",
    text="Value",
    title="Government Public Intelligence KPI",
    template="plotly_dark",
    color_continuous_scale="Viridis"
)

fig.update_traces(
    texttemplate="%{text:.2f}",
    textposition="outside"
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white",
    title_font_size=24
)

fig.show()

# Simpan screenshot sebagai:
# assets/public-intelligence-kpi.png
```

```python
# =========================================================
# STEP 5.7 — STRATEGIC RECOMMENDATIONS ENGINE
# =========================================================

recommendations = []

# Sentiment-based recommendations
if negative_rate > 20:
    recommendations.append(
        "Increase technical support and service reliability improvements."
    )

if high_urgency_rate > 10:
    recommendations.append(
        "Prioritize rapid response for high-priority public concerns."
    )

if positive_rate > 50:
    recommendations.append(
        "Leverage positive public sentiment to strengthen communication campaigns."
    )

if avg_engagement > 3000:
    recommendations.append(
        "Focus on highly engaging programs to maximize public outreach."
    )

if not recommendations:
    recommendations.append(
        "Maintain current communication and service strategies."
    )

recommendations
```

```python
# =========================================================
# STEP 5.8 — EXECUTIVE STRATEGIC INSIGHT REPORT
# =========================================================

print(f"""
============================================================
AI STRATEGIC INSIGHT REPORT
============================================================

Public Trust Index       : {public_trust_index:.2f}
Government Status        : {response_status}
Positive Sentiment       : {positive_rate:.2f}%
Negative Sentiment       : {negative_rate:.2f}%
High Urgency Issues      : {high_urgency_rate:.2f}%
Top Performing Program   : {top_program}
Top Influential Hashtag  : {top_hashtag}

Strategic Recommendations:
""")

for i, rec in enumerate(recommendations, start=1):
    print(f"{i}. {rec}")

print("""
============================================================
Strategic Insight:
The analysis indicates that public perception toward
government digital programs remains favorable. By monitoring
sentiment, urgency, and engagement, decision makers can
prioritize service improvements and optimize communication
strategies based on real-time public feedback.
============================================================
""")
```

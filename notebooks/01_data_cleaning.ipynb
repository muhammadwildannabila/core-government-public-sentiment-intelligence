STEP 1.1 — IMPORT LIBRARY

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

STEP 1.2 — LOAD DATASET

```python
DATA_PATH = "../data/raw/government_public_sentiment_dataset.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

STEP 1.3 — DATA OVERVIEW

```python
print("Dataset Shape:", df.shape)
df.info()
df.describe(include='all')
```

STEP 1.4 — CHECK MISSING VALUES

```python
missing_values = df.isnull().sum()

missing_values
```

VISUAL MISSING VALUE

```python
missing_df = pd.DataFrame({
    'column': missing_values.index,
    'missing_count': missing_values.values
})

fig = px.bar(
    missing_df,
    x='column',
    y='missing_count',
    title='Missing Value Analysis',
    template='plotly_dark',
    color='missing_count'
)

fig.update_layout(
    paper_bgcolor='#0F172A',
    plot_bgcolor='#0F172A',
    font_color='white'
)

fig.show()
```

STEP 1.5 — CHECK DUPLICATE

```python
duplicate_count = df.duplicated().sum()

print(f"Duplicate Rows: {duplicate_count}")
```

REMOVE DUPLICATE

```python
df = df.drop_duplicates()
```

STEP 1.6 — TEXT CLEANING

CLEAN TEXT FUNCTION

```python
import re
```

```python
def clean_text(text):

    text = str(text).lower()

    text = re.sub(r"http\\S+", "", text)

    text = re.sub(r"@\\w+", "", text)

    text = re.sub(r"#", "", text)

    text = re.sub(r"[^a-zA-Z\\s]", "", text)

    text = re.sub(r"\\s+", " ", text)

    return text.strip()
```

APPLY CLEANING

```python
df['clean_comment'] = df['comment_text'].apply(clean_text)
```

BEFORE VS AFTER

```python
comparison_df = df[[
    'comment_text',
    'clean_comment'
]]

comparison_df.head()
```

STEP 1.7 — SENTIMENT DISTRIBUTION

```python
sentiment_count = df['sentiment'].value_counts()

sentiment_count
```

VISUAL

```python
fig = px.pie(
    values=sentiment_count.values,
    names=sentiment_count.index,
    hole=0.6,
    title='Public Sentiment Distribution',
    template='plotly_dark',
    color_discrete_map={
        'Positive': '#22C55E',
        'Neutral': '#3B82F6',
        'Negative': '#EF4444'
    }
)

fig.update_layout(
    paper_bgcolor='#0F172A',
    plot_bgcolor='#0F172A',
    font_color='white',
    title_font_size=24
)

fig.show()
```

STEP 1.8 — CONVERT DATETIME

```python
df['created_at'] = pd.to_datetime(
    df['created_at']
)
```

EXTRACT DATE FEATURE

```python
df['year'] = df['created_at'].dt.year

df['month'] = df['created_at'].dt.month

df['day'] = df['created_at'].dt.day
```

STEP 1.9 — ENGAGEMENT CATEGORY

```python
df['engagement_category'] = pd.cut(
    df['engagement_score'],
    bins=[0, 1000, 3000, 10000],
    labels=[
        'Low',
        'Medium',
        'High'
    ]
)
```

CEK HASIL

```python
df[['engagement_score',
    'engagement_category']].head()
```

STEP 1.10 — SAVE CLEAN DATASET

```python
SAVE_PATH = "../data/processed/clean_public_sentiment.csv"

df.to_csv(
    SAVE_PATH,
    index=False
)

print("Clean dataset saved successfully.")
```

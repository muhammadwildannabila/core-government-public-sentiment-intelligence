STEP 2.1 — IMPORT LIBRARY

```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings("ignore")
```

STEP 2.2 — LOAD CLEAN DATASET

```python
DATA_PATH = "../data/processed/clean_public_sentiment.csv"

df = pd.read_csv(DATA_PATH, parse_dates=["created_at"])

print("Dataset Shape:", df.shape)
df.head()
```

STEP 2.3 — PUBLIC SENTIMENT DISTRIBUTION

```python
sentiment_count = df["sentiment"].value_counts()

fig = px.pie(
    values=sentiment_count.values,
    names=sentiment_count.index,
    hole=0.65,
    title="Public Sentiment Distribution",
    template="plotly_dark",
    color_discrete_map={
        "Positive": "#22C55E",
        "Neutral": "#3B82F6",
        "Negative": "#EF4444"
    }
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white",
    title_font_size=24
)

fig.show()
```

STEP 2.4 — SENTIMENT TREND OVER TIME

```python
daily_sentiment = (
    df.groupby([
        df["created_at"].dt.date,
        "sentiment"
    ])
    .size()
    .reset_index(name="count")
)

fig = px.line(
    daily_sentiment,
    x="created_at",
    y="count",
    color="sentiment",
    title="Public Sentiment Trend Over Time",
    template="plotly_dark",
    color_discrete_map={
        "Positive": "#22C55E",
        "Neutral": "#3B82F6",
        "Negative": "#EF4444"
    }
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white"
)

fig.show()
```

STEP 2.5 — SOCIAL MEDIA PLATFORM ANALYSIS

```python
platform_count = (
    df["platform"]
    .value_counts()
    .reset_index()
)

platform_count.columns = ["platform", "count"]

fig = px.bar(
    platform_count,
    x="platform",
    y="count",
    title="Social Media Platform Distribution",
    template="plotly_dark",
    color="count"
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white"
)

fig.show()
```

STEP 2.6 — TOP GOVERNMENT PROGRAMS BY ENGAGEMENT

```python
program_engagement = (
    df.groupby("program_name")["engagement_score"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

fig = px.bar(
    program_engagement,
    orientation="h",
    title="Top Government Programs by Engagement",
    template="plotly_dark",
    color=program_engagement.values
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white"
)

fig.show()
```

STEP 2.7 — HASHTAG INTELLIGENCE

```python
hashtag_engagement = (
    df.groupby("hashtags")["engagement_score"]
    .sum()
    .sort_values(ascending=False)
)

fig = px.bar(
    hashtag_engagement,
    title="Hashtag Intelligence Analysis",
    template="plotly_dark",
    color=hashtag_engagement.values
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white"
)

fig.show()
```

STEP 2.8 — URGENCY LEVEL MONITORING

```python
urgency_count = df["urgency_level"].value_counts()

fig = px.pie(
    values=urgency_count.values,
    names=urgency_count.index,
    hole=0.6,
    title="Urgency Level Monitoring",
    template="plotly_dark"
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    font_color="white"
)

fig.show()
```

STEP 2.9 — REGIONAL PUBLIC RESPONSE

```python
regional_engagement = (
    df.groupby("city_region")["engagement_score"]
    .sum()
    .sort_values(ascending=False)
)

fig = px.bar(
    regional_engagement,
    title="Regional Public Engagement Analysis",
    template="plotly_dark",
    color=regional_engagement.values
)

fig.update_layout(
    paper_bgcolor="#0F172A",
    plot_bgcolor="#0F172A",
    font_color="white"
)

fig.show()
```

STEP 2.10 — EXECUTIVE INSIGHT

```python
positive_rate = (
    (df["sentiment"] == "Positive").mean() * 100
)

negative_rate = (
    (df["sentiment"] == "Negative").mean() * 100
)

print(f"""
EXECUTIVE SUMMARY
-------------------------
Positive Sentiment : {positive_rate:.2f}%
Negative Sentiment : {negative_rate:.2f}%

Key Insight:
Public sentiment toward government digital programs is predominantly positive,
indicating strong acceptance and engagement from the community.
""")
```
